```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movie-basics.tsv",sep='\t')
```

```python
df.head()
```

```python
df = df.drop(columns=['titleType', 'isAdult', 'endYear'])
```

```python
df.head()
```

```python
imdb_list = pd.read_csv("tmdb_to_imdb_id_mapping.csv")
```

```python
imdb_list.head()
```

```python
imdb_list.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
imdb_list.head()
```

```python
imdb_list = imdb_list.merge(df, left_on="imdbid", right_on="tconst", how="left")
```

```python
imdb_list[imdb_list.isnull().any(axis=1)]
```

```python
imdb_list = imdb_list.dropna()
```

```python
b = pd.DataFrame(imdb_list['genres'].str.split(',').tolist(), index=imdb_list.tconst).stack()
b = b.reset_index()[[0, 'tconst']] # var1 variable is currently labeled 0
b.columns = ['genre', 'imdbid'] # renaming var1
```

```python
b
```

```python
cols = b.columns.tolist()
cols = cols[-1:] + cols[:-1]
cols
b = b[cols]
```
