```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("imdb-cast.tsv",sep='\t')
```

```python
df = df[(df['category']=='actor') | (df['category']=='director')]
```

```python
df.head(20)
```

```python
df = df.drop(columns=['job','characters', 'ordering'])
```

```python
df.head()
```

```python
imdb_list = pd.read_csv("tmdb_to_imdb_id_mapping.csv")
imdb_list = imdb_list.drop(columns="Unnamed: 0")
```

```python
imdb_list.head()
```

```python
df = df.merge(imdb_list, left_on="tconst", right_on="imdbid", how="left")
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df['tmdbid'] = df['tmdbid'].apply(lambda x: int(x))
df.head()
```

```python
df.to_csv("filtered_cast_list.csv")
```
