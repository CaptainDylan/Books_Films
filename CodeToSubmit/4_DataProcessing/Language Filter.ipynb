
## Use langdetect to filter out non-English reviews

```python
# Do this first
!pip install langdetect
```

```python
from langdetect import detect
```

### We have to write a wrapper because if the text is too short detect() throws an error :-(

```python
def getlang(str):
    try:
        result = detect(str)
    except:
        result = "??"
    return result
```

```python
goodreadsreviews['language'] = goodreadsreviews['review_text'].apply(getlang)
#detect(goodreadsreviews.iloc[994]['review_text'])
```

```python
goodreadsreviews[goodreadsreviews['language']!='en']
```
